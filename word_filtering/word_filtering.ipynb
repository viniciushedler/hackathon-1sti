```python
bottom_limit = 3
top_limit = 10

blocked_words = []

with open("valid_words.txt", "r") as f:
    new_valid_words = []
    for line in f.readlines():
        len_line = len(line)-1 # -1 because of the \n token
        if len_line>bottom_limit and len_line<top_limit:
            new_valid_words.append(line)
        else:
            blocked_words.append(line)

with open("new_valid_words.txt", "w") as f:
    f.writelines(new_valid_words)

blocked_words
```

```python
import secrets
```

```python
def get_random_word():
    word_f_address = "../word_filtering/new_valid_words.txt"
    with open(word_f_address, 'r') as f:
        word = secrets.choice(f.readlines())
        word = word[:-1]
    return word

get_random_word()
```
